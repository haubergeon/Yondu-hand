# Whistle Control for Robotic Hand


```python
import pyaudio
import struct
import numpy as np
import matplotlib.pyplot as plt
%matplotlib tk
```

```python
CHUNK = 1024 * 4
# audio samples per frame
FORMAT = pyaudio.paInt16

CHANNELS = 1

RATE = 44100
```

```python
p = pyaudio.PyAudio()

stream = p.open(
format = FORMAT,
channels = CHANNELS,
rate = RATE,
input = True,
output = True,
frames_per_buffer = CHUNK
)

data = stream.read(CHUNK)
# this returns data in a weird byte like format so to convert it to integers we use struct

data_int = struct.unpack(str(2*CHUNK) +'B', data)

data_int
```

```python
fig, ax = plt.subplots()
ax.plot(data_int)
plt.show
```
